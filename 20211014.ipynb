```python
conditions = [
              (df['SEX'] == 1) & (df['GAMMA_GTP']<11) & (df['SGOT_AST']>40) & (df['SGPT_ALT']>40),
              (df['SEX'] == 1) & (df['GAMMA_GTP']>63) & (df['SGOT_AST']>40) & (df['SGPT_ALT']>40),
              (df['SEX'] == 2) & (df['GAMMA_GTP']<8) & (df['SGOT_AST']>40) & (df['SGPT_ALT']>40),
              (df['SEX'] == 2) & (df['GAMMA_GTP']>35) & (df['SGOT_AST']>40) & (df['SGPT_ALT']>40)
]
vals = [1, 1, 1, 1]

df['liver'] = np.select(conditions, vals)
```
